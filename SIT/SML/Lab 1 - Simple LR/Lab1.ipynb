```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Import Dataset

```python
dataset = None

with open("./dataset.txt", "r") as file:
    data = file.read()
    splitted = data.split()
    data_extracted = [[],[]]
    switch = 0
    for i in range(len(splitted)):
        if(switch == 0):
            data_extracted[0].append(float(splitted[i]))
            switch = 1
        else:
            data_extracted[1].append(float(splitted[i]))
            switch = 0
    dataset = np.array(data_extracted)
```

```python
data = pd.DataFrame()
data["X"], data["Y"] = dataset[0], dataset[1]
```

```python
plt.title("Advertisement vs Sales")
plt.scatter(x=data.X, y=data.Y)
plt.legend(["Sales vs advertising"])
plt.xlabel("Advertisement"), plt.ylabel("Sales")
plt.tight_layout()
```

## Simple Linear Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

xtrain, xtest, ytrain, ytest = train_test_split(data["X"], data["Y"], test_size=0.33, random_state=42)

xtrain, xtest, ytrain, ytest = np.array(xtrain).reshape(-1, 1), np.array(xtest).reshape(-1, 1), np.array(ytrain).reshape(-1, 1), np.array(ytest).reshape(-1, 1)
```

```python
model = LinearRegression()
model.fit(xtrain, ytrain)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

ypred = model.predict(xtest)

print(f"weight: {model.coef_[0][0]}")
print(f"bias: {model.intercept_[0]}")

mse = mean_squared_error(y_true=ytest, y_pred=ypred)

print(f"mse: {mse}")
print(f"rmse: {np.sqrt(mean_squared_error(y_true=ytest, y_pred=ypred))}")
print(f"r2: {r2_score(y_true=ytest, y_pred=ypred)}")
```

```python
plt.title("Advertisement vs Sales Regression")
plt.scatter(x=data.X, y=data.Y)
plt.plot(xtest, ypred)
plt.legend(["Sales vs advertising"])
plt.xlabel("Advertisement"), plt.ylabel("Sales")
plt.tight_layout()
```

## Regression on Salary Data

```python
data = pd.read_csv("./Salary_Data.csv")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

xtrain, xtest, ytrain, ytest = train_test_split(data["YearsExperience"], data["Salary"], test_size=0.33, random_state=42)

xtrain, xtest, ytrain, ytest = np.array(xtrain).reshape(-1, 1), np.array(xtest).reshape(-1, 1), np.array(ytrain).reshape(-1, 1), np.array(ytest).reshape(-1, 1)
```

```python
model = LinearRegression()
model.fit(xtrain, ytrain)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

ypred = model.predict(xtest)

print(f"weight: {model.coef_[0][0]}")
print(f"bias: {model.intercept_[0]}")

mse = mean_squared_error(y_true=ytest, y_pred=ypred)

print(f"mse: {mse}")
print(f"rmse: {np.sqrt(mean_squared_error(y_true=ytest, y_pred=ypred))}")
print(f"r2: {r2_score(y_true=ytest, y_pred=ypred)}")
```
