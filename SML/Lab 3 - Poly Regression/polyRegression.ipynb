# Polynomial Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

##

## Lab Dataset

```python
data = pd.read_excel("./data_.xlsx")
data
```

#### Fitting Linear Regression Model

```python
X = np.array(data["Temperature"]).reshape(-1, 1)
y = np.array(data["Pressure"]).reshape(-1, 1)
```

```python
LR_model = LinearRegression()
LR_model.fit(X=X, y=y)
LR_pred = LR_model.predict(X=X)
```

#### Fitting Polynomial Regression Model

```python
poly_features = PolynomialFeatures(degree = 4)
poly_x = poly_features.fit_transform(X)
```

```python
lr = LinearRegression().fit(poly_x, y)
poly_pred = lr.predict(poly_x)

plt.xlabel("Temperature")
plt.ylabel("Pressure")
plt.scatter(X, y, color="Black", label="Ground Truth")
plt.plot(X, LR_pred, color="red", label="Simple Linear Regression")
plt.plot(X, poly_pred, color="green", label="Polynomial Regression")

plt.legend()
```

```python
print(lr.coef_, lr.intercept_, sep="\n\n")
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# print(f"weight: {lr.coef_[0][0]}")
# print(f"bias: {lr.intercept_[0]}")

mse = mean_squared_error(y_true=y, y_pred=poly_pred)

print(f"mse: {mse}")
print(f"rmse: {np.sqrt(mean_squared_error(y_true=y, y_pred=poly_pred))}")
print(f"r2: {r2_score(y_true=y, y_pred=poly_pred)}")
```
